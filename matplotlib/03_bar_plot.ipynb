# Bar Plot in Matplotlib

A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally.

## 1. Creating a Simple Bar Chart
Using `plt.bar()` to create vertical bars.

```python
import matplotlib.pyplot as plt

# Creating the dataset
data = {'C': 20, 'C++': 15, 'Java': 30, 'Python': 35}
courses = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize=(10, 5))

# Creating the bar plot
plt.bar(courses, values, color='maroon', width=0.4)

plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```

## 2. Horizontal Bar Chart
Using `plt.barh()` to create horizontal bars. This is useful when category names are long.

```python
import matplotlib.pyplot as plt

courses = ['C', 'C++', 'Java', 'Python']
values = [20, 15, 30, 35]

plt.barh(courses, values, color='skyblue')

plt.xlabel("No. of students enrolled")
plt.ylabel("Courses offered")
plt.title("Horizontal Bar Chart")
plt.show()
```

## 3. Grouped Bar Chart
To compare multiple sets of data, we can offset the x-positions of the bars.

```python
import numpy as np
import matplotlib.pyplot as plt

IT = [12, 30, 1, 8, 22]
ECE = [28, 6, 16, 5, 10]
CSE = [29, 3, 24, 25, 17]

barWidth = 0.25
fig = plt.subplots(figsize=(12, 8))

# Set position of bar on X axis
br1 = np.arange(len(IT))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

# Make the plot
plt.bar(br1, IT, color='r', width=barWidth, edgecolor='grey', label='IT')
plt.bar(br2, ECE, color='g', width=barWidth, edgecolor='grey', label='ECE')
plt.bar(br3, CSE, color='b', width=barWidth, edgecolor='grey', label='CSE')

# Adding xticks
plt.xlabel('Branch', fontweight='bold', fontsize=15)
plt.ylabel('Students', fontweight='bold', fontsize=15)
plt.xticks([r + barWidth for r in range(len(IT))], ['2015', '2016', '2017', '2018', '2019'])

plt.legend()
plt.show()
```

## 4. Stacked Bar Chart
Stacked bar charts represent different groups on top of one another.

```python
import matplotlib.pyplot as plt

x = ['A', 'B', 'C', 'D']
y1 = [10, 20, 10, 30]
y2 = [20, 25, 15, 25]

# plot bars in stack manner
plt.bar(x, y1, color='r')
plt.bar(x, y2, bottom=y1, color='b')
plt.show()
```
