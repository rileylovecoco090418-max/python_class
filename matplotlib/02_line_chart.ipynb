# Line Chart in Matplotlib - Python

A line chart or line plot is a graphical representation used to show the relationship between two continuous variables by connecting data points with a straight line. It is commonly used to visualize trends, patterns, or changes over time.

## 1. Simple Line Plot
A basic line chart connecting data points on the x and y axes.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y = x * 2

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.show()
```

## 2. Adding Labels, Title, and Markers
Markers help highlight individual data points, and labels make the chart readable.

```python
import matplotlib.pyplot as plt
import numpy as np

days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
temperature = [22, 24, 23, 26, 25]

plt.plot(days, temperature, marker='o', linestyle='-', color='b', label='Temp')
plt.title('Weekly Temperature')
plt.xlabel('Days')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.show()
```

## 3. Adding a Grid
Grids make it easier to read values across the chart.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.grid(True)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot with Grid")
plt.show()
```

## 4. Line Chart with Annotations
Using `annotate()` to display exact values on data points.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o', linestyle='-')

for xi, yi in zip(x, y):
    plt.annotate(f'({xi}, {yi})', (xi, yi), textcoords="offset points", xytext=(0, 10), ha='center')

plt.title('Line Chart with Annotations')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

## 5. Multiple Plots on the Same Axis
Comparing different datasets within the same coordinate system.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y = x * 2
x1 = [1, 2, 3, 4]
y1 = [3, 5, 7, 9]

plt.plot(x, y, label='y = 2x')
plt.plot(x1, y1, '-.', label='Second Line')
plt.xlabel("X-axis data")
plt.ylabel("Y-axis data")
plt.title("Multiple Plots on Same Axis")
plt.legend()
plt.show()
```

## 6. Fill the Area Between Two Lines
Using `fill_between()` to shade the region between two curves.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y = x * 2
y1 = [3, 5, 7, 9]

plt.plot(x, y, label='y = 2x')
plt.plot(x, y1, '-.', label='y1')
plt.fill_between(x, y, y1, color='green', alpha=0.4)

plt.xlabel("X-axis data")
plt.ylabel("Y-axis data")
plt.title("Filled Area Between Two Lines")
plt.legend()
plt.show()
```

## 7. Plotting Trigonometric Functions
Visualizing sine and cosine waves using `np.linspace()`.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, label="sin(x)")
plt.plot(x, y_cos, label="cos(x)")

plt.xlabel("X values")
plt.ylabel("Function value")
plt.title("Trigonometric Functions")
plt.legend()
plt.show()
```

## 8. Saving a Plot
Save your chart as an image file using `savefig()`.

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
plt.plot(x, y)
plt.title("Saved Line Plot")
plt.savefig("line_plot.png")
plt.show()
```
